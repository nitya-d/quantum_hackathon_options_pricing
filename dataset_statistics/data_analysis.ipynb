```python
from datasets import load_dataset

ds_level2 = load_dataset(
    "Quandela/Challenge_Swaptions",
    data_files="level-2_Missing_data_prediction/train_level2.csv",
    split="train",
)
print(ds_level2)
```

```python
df1 = ds_level1
df2 = ds_level2
```

```python
import pandas as pd
```

```python
df1 = ds_level1.to_pandas()
df1
```

```python
df2 =ds_level2.to_pandas()
df2
```

```python
df1['Date'] = pd.to_datetime(df1['Date'], dayfirst=True)
```

```python
df1 = df1.sort_values('Date').reset_index(drop=True)

print(f"Date range   : {df1['Date'].min().date()}  →  {df1['Date'].max().date()}")
print(f"Total rows   : {len(df1)}")
print(f"Total columns: {df1.shape[1]}  (1 Date + {df1.shape[1]-1} swaption values)")
```

```python
feature_cols = [c for c in df1.columns if c != 'Date']

# Extract unique tenors and maturities
tenors    = sorted(set(int(c.split('Tenor : ')[1].split(';')[0]) for c in feature_cols))
maturities = sorted(set(float(c.split('Maturity : ')[1]) for c in feature_cols))

print(f"\nTenors     ({len(tenors)}): {tenors}")
print(f"Maturities ({len(maturities)}): {[round(m,4) for m in maturities]}")
```

```python
# ─────────────────────────────────────────────
# 3. MISSING VALUES
# ─────────────────────────────────────────────

print("\n" + "=" * 60)
print("MISSING VALUES")
print("=" * 60)

missing1 = df1[feature_cols].isnull().sum().sum()
missing2 = df2[[c for c in df2.columns if c != 'Date']].isnull().sum().sum()
total_cells2 = df2.shape[0] * (df2.shape[1] - 1)

print(f"Level 1 — missing values: {missing1}")
print(f"Level 2 — missing values: {missing2}  ({100*missing2/total_cells2:.1f}% of all cells)")
```

```python
import numpy as np
```

```python
# ─────────────────────────────────────────────
# 4. SUMMARY STATISTICS
# ─────────────────────────────────────────────

print("\n" + "=" * 60)
print("SUMMARY STATISTICS (Level 1 swaption values)")
print("=" * 60)

vals = df1[feature_cols].values.flatten()
vals = vals[~np.isnan(vals)]

print(f"  Min    : {vals.min():.6f}")
print(f"  Max    : {vals.max():.6f}")
print(f"  Mean   : {vals.mean():.6f}")
print(f"  Std    : {vals.std():.6f}")
print(f"  Median : {np.median(vals):.6f}")

# ─────────────────────────────────────────────
# 5. HELPER: build surface matrix for a given row
# ─────────────────────────────────────────────

def build_surface(row, tenors, maturities, feature_cols):
    """Return a (len(maturities) x len(tenors)) matrix for one row."""
    mat = np.full((len(maturities), len(tenors)), np.nan)
    for i, mat_val in enumerate(maturities):
        for j, ten_val in enumerate(tenors):
            col = f"Tenor : {ten_val}; Maturity : {mat_val}"
            if col in row.index:
                mat[i, j] = row[col]
    return mat
```

```python
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
import seaborn as sns
```

```python
# ─────────────────────────────────────────────
# 6. PLOTS
# ─────────────────────────────────────────────

fig = plt.figure(figsize=(22, 26))
fig.patch.set_facecolor('#0d1117')
gs = gridspec.GridSpec(4, 2, figure=fig, hspace=0.45, wspace=0.35)

CMAP      = 'plasma'
TEXT_COL  = '#e6edf3'
GRID_COL  = '#21262d'
ACCENT    = '#58a6ff'
ACCENT2   = '#f78166'

plt.rcParams.update({
    'text.color': TEXT_COL,
    'axes.labelcolor': TEXT_COL,
    'xtick.color': TEXT_COL,
    'ytick.color': TEXT_COL,
    'axes.facecolor': '#161b22',
    'axes.edgecolor': GRID_COL,
    'grid.color': GRID_COL,
    'figure.facecolor': '#0d1117',
})

# ── 6a. Average volatility surface (Level 1)
ax1 = fig.add_subplot(gs[0, 0])
avg_surface = np.nanmean(
    np.stack([build_surface(df1.iloc[i], tenors, maturities, feature_cols) for i in range(len(df1))]),
    axis=0
)
im = ax1.imshow(avg_surface, cmap=CMAP, aspect='auto', origin='lower')
ax1.set_xticks(range(len(tenors)));    ax1.set_xticklabels(tenors, fontsize=8)
ax1.set_yticks(range(len(maturities))); ax1.set_yticklabels([round(m,2) for m in maturities], fontsize=8)
ax1.set_xlabel('Tenor (years)');  ax1.set_ylabel('Maturity (years)')
ax1.set_title('Average Swaption Surface — Level 1', color=TEXT_COL, fontsize=12, pad=10)
plt.colorbar(im, ax=ax1, fraction=0.046)

# ── 6b. Volatility distribution
ax2 = fig.add_subplot(gs[0, 1])
ax2.hist(vals, bins=80, color=ACCENT, edgecolor='none', alpha=0.85)
ax2.axvline(vals.mean(), color=ACCENT2, lw=2, linestyle='--', label=f'Mean={vals.mean():.4f}')
ax2.set_xlabel('Implied Volatility');  ax2.set_ylabel('Count')
ax2.set_title('Distribution of All Swaption Values', color=TEXT_COL, fontsize=12)
ax2.legend(facecolor='#161b22', labelcolor=TEXT_COL)
ax2.grid(True, alpha=0.3)

# ── 6c. Time series of a few key cells
ax3 = fig.add_subplot(gs[1, :])
sample_cols = [
    'Tenor : 5; Maturity : 1',
    'Tenor : 10; Maturity : 5',
    'Tenor : 2; Maturity : 10',
    'Tenor : 30; Maturity : 30',
]
colors_ts = [ACCENT, ACCENT2, '#3fb950', '#d2a8ff']
for col, color in zip(sample_cols, colors_ts):
    if col in df1.columns:
        label = col.replace('Tenor : ', 'T').replace('; Maturity : ', '/M')
        ax3.plot(df1['Date'], df1[col], color=color, lw=1.2, label=label, alpha=0.9)
ax3.set_title('Time Series — Selected Swaption Points', color=TEXT_COL, fontsize=12)
ax3.set_xlabel('Date');  ax3.set_ylabel('Implied Volatility')
ax3.legend(facecolor='#161b22', labelcolor=TEXT_COL, ncol=4)
ax3.grid(True, alpha=0.3)

# ── 6d. Correlation heatmap (sample of tenors at Maturity=1)
ax4 = fig.add_subplot(gs[2, 0])
mat1_cols = [c for c in feature_cols if 'Maturity : 1' in c and c.endswith('; Maturity : 1') == False]
mat1_cols = [c for c in feature_cols if c.split('Maturity : ')[1] == '1']
if len(mat1_cols) > 1:
    corr = df1[mat1_cols].corr()
    short_labels = [f"T{c.split('Tenor : ')[1].split(';')[0]}" for c in mat1_cols]
    sns.heatmap(corr, ax=ax4, cmap='coolwarm', annot=True, fmt='.2f',
                xticklabels=short_labels, yticklabels=short_labels,
                annot_kws={'size': 7}, linewidths=0.3)
    ax4.set_title('Tenor Correlation (Maturity = 1yr)', color=TEXT_COL, fontsize=12)
    ax4.tick_params(colors=TEXT_COL)

# ── 6e. Missing values heatmap — Level 2
ax5 = fig.add_subplot(gs[2, 1])
feat2 = [c for c in df2.columns if c != 'Date']
missing_per_col = df2[feat2].isnull().mean()
# reshape into surface
miss_mat = np.full((len(maturities), len(tenors)), np.nan)
for i, mat_val in enumerate(maturities):
    for j, ten_val in enumerate(tenors):
        col = f"Tenor : {ten_val}; Maturity : {mat_val}"
        if col in missing_per_col.index:
            miss_mat[i, j] = missing_per_col[col]
im5 = ax5.imshow(miss_mat, cmap='Reds', aspect='auto', origin='lower', vmin=0, vmax=1)
ax5.set_xticks(range(len(tenors)));    ax5.set_xticklabels(tenors, fontsize=8)
ax5.set_yticks(range(len(maturities))); ax5.set_yticklabels([round(m,2) for m in maturities], fontsize=8)
ax5.set_xlabel('Tenor');  ax5.set_ylabel('Maturity')
ax5.set_title('Missing Rate by Cell — Level 2', color=TEXT_COL, fontsize=12)
plt.colorbar(im5, ax=ax5, fraction=0.046, label='Fraction missing')

# ── 6f. Day-to-day changes (volatility of volatility)
ax6 = fig.add_subplot(gs[3, :])
daily_mean = df1[feature_cols].mean(axis=1)
daily_change = daily_mean.diff().dropna()
ax6.fill_between(df1['Date'].iloc[1:], daily_change, 0,
                  where=daily_change >= 0, color=ACCENT,  alpha=0.7, label='Up')
ax6.fill_between(df1['Date'].iloc[1:], daily_change, 0,
                  where=daily_change < 0,  color=ACCENT2, alpha=0.7, label='Down')
ax6.axhline(0, color=TEXT_COL, lw=0.5)
ax6.set_title('Day-to-Day Change in Average Surface Volatility', color=TEXT_COL, fontsize=12)
ax6.set_xlabel('Date');  ax6.set_ylabel('Δ Avg Volatility')
ax6.legend(facecolor='#161b22', labelcolor=TEXT_COL)
ax6.grid(True, alpha=0.3)

fig.suptitle('Swaption Dataset — Exploratory Analysis', 
             fontsize=16, color=TEXT_COL, y=1.01, fontweight='bold')

plt.savefig('swaption_analysis.png', 
            dpi=150, bbox_inches='tight', facecolor='#0d1117')
print("\nPlot saved → swaption_analysis.png")

# ─────────────────────────────────────────────
# 7. LEVEL 1 — AUTO-CORRELATION CHECK
# ─────────────────────────────────────────────

print("\n" + "=" * 60)
print("AUTO-CORRELATION (Level 1) — useful for time series modeling")
print("=" * 60)

# Check lag-1 autocorrelation for each feature
autocorrs = df1[feature_cols].apply(lambda x: x.autocorr(lag=1))
print(f"  Mean lag-1 autocorrelation : {autocorrs.mean():.4f}")
print(f"  Min  lag-1 autocorrelation : {autocorrs.min():.4f}")
print(f"  Max  lag-1 autocorrelation : {autocorrs.max():.4f}")
print("  → High autocorrelation means yesterday's surface is a strong predictor of today's.")
```

```python
# ─────────────────────────────────────────────
# 8. LEVEL 2 — MISSING PATTERN ANALYSIS
# ─────────────────────────────────────────────

print("\n" + "=" * 60)
print("MISSING DATA PATTERN ANALYSIS (Level 2)")
print("=" * 60)

feat2 = [c for c in df2.columns if c != 'Date']
missing_per_row = df2[feat2].isnull().sum(axis=1)
print(f"  Rows with NO missing values  : {(missing_per_row == 0).sum()}")
print(f"  Rows with SOME missing values: {(missing_per_row > 0).sum()}")
print(f"  Avg missing per row          : {missing_per_row.mean():.1f} / {len(feat2)} cells")
print(f"  Max missing in a single row  : {missing_per_row.max()}")

missing_per_col_pct = df2[feat2].isnull().mean().sort_values(ascending=False)
print(f"\n  Top 5 most-missing columns:")
for col, pct in missing_per_col_pct.head(5).items():
    short = col.replace('Tenor : ', 'T').replace('; Maturity : ', '/M')
    print(f"    {short:<15} → {pct*100:.1f}% missing")

print("\nDone! All outputs saved.")
```
